```python
# Check missing values in the dataset and find the appropriate measures to fill in the missing values
data.isna().sum()
```

```python
# For numerical columns
```

```python
data['bmi'] = data['bmi'].fillna(data['bmi'].median())
```

```python
# For categorical columns
```

```python
data['region'] = data['region'].fillna(data['region'].mode()[0])
```

```python
sns.pairplot (x_vars='charges',y_vars=['age','bmi','children'],data=data)
plt.show()
```

```python
Observation:

This dataset usually has zero missing values

If found, median/mode imputation is appropriate.
```

```python
# Explore the relationship between the feature and target column using a count plot of categorical columns and a scatter plot of numerical columns
# Scatter Plots for Numerical Columns
fig,axes=plt.subplots(1,3,figsize=(15,5))
axes[0].scatter('charges','age',data=data,color='red')
axes[0].set_title('scatterplot1')
axes[1].scatter('charges','bmi',data=data,color='blue')
axes[1].set_title('scatterplot2')
axes[2].scatter('charges','children',data=data,color='green')
axes[2].set_title('scatterplot3')
plt.show()
```

```python
categorical_cols=['sex','smoker','region']
```

```python
categorical_cols
```

```python
# Count Plots for Categorical Columns
categorical_cols=['sex','smoker','region']
fig,axes=plt.subplots(1,3,figsize=(15,5))
for i, col in enumerate(categorical_cols):
    sns.countplot (x=col,data=data,ax=axes[i])
    axes[i].set_title(f"countplot of{col}")
plt.show()
```

```python
Observations:

Smokers are fewer in number but show very high charges

Charges increase as age increases

BMI above 30 shows noticeably higher charges

Region has no major effect on charges
```

```python
# Check if the number of premium charges for smokers or non-smokers is increasing as they are aging
sns.scatterplot(x='age',y='charges',hue='smoker',data=data)
plt.show()
```

```python
Observation:

Non-smokers: charges increase slowly with age

Smokers: charges are extremely high and rise sharply with age

Smoking has the strongest effect on premium charges.
```

```python
# Perform data visualization using plots of feature vs feature
plt.figure(figsize=(8,5))
sns.heatmap(data.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

```python
Observation:

Highest correlation: smoker - charges

Age also correlates positively with charges

Children has low correlation
```
